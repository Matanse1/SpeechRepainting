```python
import torch
import torch.nn.functional as F

def test_attention():
    batch_size, q_seq_len, embed_dim, heads = 2, 10, 8, 2
    k_seq_len = 10
    
    query = torch.randn(batch_size, heads, q_seq_len, embed_dim)
    key = torch.randn(batch_size, heads, k_seq_len, embed_dim)
    value = torch.randn(batch_size, heads, k_seq_len, embed_dim)
    
    # Test without mask
    x_no_mask = F.scaled_dot_product_attention(query, key, value, attn_mask=None, dropout_p=0.0, is_causal=False)
    # print("Output without mask:\n", x_no_mask)
    
    # Test with mask
        # Test with mask
    mask = torch.zeros(batch_size, q_seq_len, dtype=torch.bool)
    for i in range(batch_size):
        rand_idx = torch.randint(q_seq_len-10, q_seq_len, (1,)).item()
        print("Random index:", rand_idx)
        mask[i, rand_idx:] = True
    print("\nMask:\n", mask)
    #my mask
    attn_mask = mask.unsqueeze(2) | mask.unsqueeze(1)  # [B, T1, T2]
    attn_mask = attn_mask.unsqueeze(1).expand(-1, heads, -1, -1)  # [B, n_head, T1, T2]
    attn_mask = ~attn_mask
    print("\nAttn Mask:\n", attn_mask)
    # attn_mask = attn_mask.permute(1, 0, 2, 3).contiguous().view(-1, q_seq_len, k_seq_len) 

    # in paper
    # attn_mask = mask.unsqueeze(1).unsqueeze(1)  # 'b n -> b 1 1 n'
    # attn_mask = attn_mask.expand(batch_size, heads, q_seq_len, k_seq_len)
    
    x_with_mask = F.scaled_dot_product_attention(query, key, value, attn_mask=attn_mask, dropout_p=0.0, is_causal=False)
    print("\nOutput with mask:\n", x_with_mask)

if __name__ == "__main__":
    test_attention()
```
